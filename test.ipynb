# BDT

```python
import random,math,hashlib
def trailing_zeros(num):
    if num == 0:
        return 64
    p = 0
    while(num>>p) & 1 == 0:
        p += 1
    return p 
def estimate_cardinality(values):
    my_max = 0
    for value in values:
        digest = hashlib.sha1(str(value).encode("utf-8")).hexdigest()
        h = int(digest,16)
        zeros = trailing_zeros(h)
        if(zeros > my_max):
            my_max = zeros
#         print(value,"\t",bin(h)[2:].zfill(160),"\t T: ",trailing_zeros(h))
    return math.pow(2,my_max)
```

```python
k = 10
repetitions = 100
sum_rel = 0
for j in range(repetitions):
    mylist = [int(random.gauss(100,5)) for i in range(k)]
#     print(mylist)
    true_distinct = len(set(mylist))
    approximate_distinct = estimate_cardinality(mylist)
    print("True distinct: ", true_distinct ,"Estimate: ",approximate_distinct)
    rel=abs(true_distinct-approximate_distinct)/true_distinct*100
    sum_rel += rel
print(sum_rel/float(repetitions))
```

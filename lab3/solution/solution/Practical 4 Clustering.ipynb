# Write K-Means Yourself

```python
import random
import numpy as np
import sklearn.datasets as data
import matplotlib.pyplot as plt

DEBUGGING = False
PLOTTING = False
# PLOTS_DIR = '../plots/'
# define markers for up to 10 clusters
CLUSTER_MARKERS = [ 'bo', 'rv', 'c^', 'm<', 'y>', 'ks', 'bp', 'r*', 'cD', 'mP' ]

#-generate synthetic data for clustering
X ,clusters = data.make_blobs( n_samples=100, n_features=2, cluster_std=2, random_state=2022 )

M = len(X)
minX0 = np.min( X[:,0] )
maxX0 = np.max( X[:,0] )
minX1 = np.min( X[:,1] )
maxX1 = np.max( X[:,1] )

#-(optionally) print some info about the data set
if DEBUGGING:
    print('number of instances = ' + str( M ))
    print('X[:,0] min={} max={}   X[:,1] min={} max={}'.format( minX0, maxX0, minX1, maxX1 ))
    
#-plot raw data --- always a good idea to do this!
if PLOTTING:
    plt.figure()
    # plot data points
    plt.plot( X[:,0], X[:,1], 'g.', markersize=10 )
    plt.show()
    plt.close()
    
#-loop through various numbers of clusters
BC = np.zeros( 10 )
WC = np.zeros( 10 )
for K in range( 2, 10 ):
    #-initialise clusters by finding random points for K cluster centres,
    # using the minimum and maximum values in the data set
    centres = np.zeros(( K, 2 ))
    for i in range(K):
        centres[i][0] = random.uniform(minX0,maxX0)
        centres[i][1] = random.uniform(minX1,maxX1)
    
    dist    = np.zeros(( M, K )) # for each instance, distance to each cluster centre
    labels  = np.zeros( M, dtype=int ) # for each instance, label of cluster with closest centre
    members = [[] for i in range (K)]
    converged = False
    iters = 0
    while(not converged):
        converged = True
        for j in range (M):
            for i in range(K):
                dist[j,i] = np.sqrt(np.square(X[j,0] - centres[i,0])+ np.square(X[j,1]- centres[i,1]))
            this_cluster = np.argmin(dist[j,:])
            if(labels[j] != this_cluster):
                converged = False
                labels[j] = this_cluster #dont want this indent！！！！！！！！！！！！！！！！！！！！！！！！11
            members[this_cluster].append( j )
            
        #-plot clusters
        if PLOTTING and (iters% 5 == 0 or converged) :
            plt.figure()
            for i in range(M):
                plt.plot(X[i,0],X[i,1],CLUSTER_MARKERS[labels[i]], markersize = 10)
                
            plt.xlabel("x[0]",fontsize = 14)
            plt.ylabel("x[1]",fontsize = 14)
            plt.title("cluster: "+str(K)+" iters: "+str(iters),fontsize = 14)
            plt.show();
            plt.close()
           
        #-compute new cluster centres
        for i in range( K ):
            if(len(members[i]) > 0):
                centres[i,0] = 0
                centres[i,1] = 0
                for m in range(len(members[i])):
                    centres[i,0] += X[members[i][m],0]
                    centres[i,1] += X[members[i][m],1]
                centres[i,0] /= len( members[i] )
                centres[i,1] /= len( members[i] )
            else: # no need ?????????????????????????
                centres[i,0] = random.uniform( minX0, maxX0 )
                centres[i,1] = random.uniform( minX1, maxX1 )
        iters += 1
        
        
        within = np.zeros(K)
        for i in range(K):
            for j in (members[i]):
                within[i] += np.square(X[j,0] - centres[i,0]) + np.square(X[j,1] - centres[i,1])
        WC[K] = np.sum(within)
        
        between = np.zeros(K)
        for i in range(K):
            for j in range(i+1,K):
                between[i] = np.square(centres[i,0] - centres[j,0]) + np.square(centres[i,1] - centres[j,1])
        BC[K] = np.sum(between)
        
        score = BC[K]/WC[K]
        print('K={}  WC={}  BC={}  score={} iters ={}'.format( K, WC[K], BC[K], score,iters ))
#-plot overall scores
plt.figure()
plt.plot( BC, linewidth=2, label='Between Cluster' )
plt.plot( WC, linewidth=2, label='Within Cluster' )
plt.xlabel( 'K' )
plt.legend( loc='best' )
plt.show()
plt.close()
```

# Use K-Means from scikit-learn

```python
import numpy as np
import sklearn.datasets as data
import matplotlib.pyplot as plt
import sklearn.cluster as cluster
import sklearn.metrics as metrics

DEBUGGING = False
# PLOTS_DIR = '../plots/'

# define markers for up to 10 clusters
CLUSTER_MARKERS = [ 'bo', 'rv', 'c^', 'm<', 'y>', 'ks', 'bp', 'r*', 'cD', 'mP' ]

X, cluters = data.make_blobs( n_samples=1000, n_features=2, cluster_std=2, random_state=2023)
# set number of instances
M = len(X)

plt.figure()
plt.plot(X[:,0],X[:,1],"g.",markersize = 10)
plt.show()
plt.close()

#-loop through various numbers of clusters
BC = np.zeros( 10 ) # between cluster
WC = np.zeros( 10 ) # within cluster
IN = np.zeros( 10 ) # inertia (within cluster)
SC = np.zeros( 10 ) # silhouette coefficient
CH = np.zeros( 10 ) # calinksi-harabasz

K = 5

km = cluster.KMeans(n_clusters= K)

km.fit(X)

# record inertia score (within clusters, computed by scikit function)
IN[K] = km.inertia_

#-compute silhouette score
SC[K] = metrics.silhouette_score(X, km.labels_, metric='euclidean' )

#-compute calinski-harabasz score
CH[K] = metrics.calinski_harabasz_score(X,km.labels_)

#-tally members of each cluster
# members = [[] for i in range (K)]
# for i in range(K):
#     for j in range(M):
#         if(km.labels_[j] == i):
#             members[i].append(j)
members = [[] for i in range( K )] # lists of members of each cluster
for j in range( M ): # loop through instances
    members[ km.labels_[j] ].append( j ) # add this instance to cluster returned by scikit function

within = np.zeros( K )
for i in range(K):
    for j in members[i]:
        within[i] += np.square(X[j,0] - km.cluster_centers_[i][0])+ np.square(X[j,1] - km.cluster_centers_[i][1])
WC[K] = np.sum(within)

between = np.zeros( K )
for i in range (K):
    for j in range(i+1, K):
        between[i] += np.square(km.cluster_centers_[i][0] - km.cluster_centers_[j][0]) 
        + np.square(km.cluster_centers_[i][1] - km.cluster_centers_[j][1])
BC[K] = np.sum( between )

score = BC[K] / WC[K]

print('K={}  WC={}  BC={}  score={}  inertia={}  silhouette={}  calinski-harabasz={}'.format( K, WC[K], BC[K], score, IN[K], SC[K], CH[K] ))

plt.figure()
for i in range (M):
    plt.plot(X[i,0],X[i,1],CLUSTER_MARKERS[km.labels_[i]])
plt.xlabel( 'X[0]' )
plt.ylabel( 'X[1]' )
plt.title( 'k-means clustering, K=' + str( K ))
plt.show()
plt.close()

plt.figure()
a, = plt.plot( BC, linewidth=2 )
b, = plt.plot( WC, linewidth=2 )
c, = plt.plot( SC*1000, linewidth=2 )
d, = plt.plot( CH, linewidth=2 )
plt.xlabel( 'K' )
plt.legend( (a,b,c,d), ('Between Cluster','Within Cluster','Silhouette Coefficient(x1K)','Calinski-Harabasz' ),loc='best' )
plt.show()
plt.close()
```
